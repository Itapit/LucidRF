```python
import h5py
import numpy as np
```

```python
def create_master_split():
    try:
        f = h5py.File(RAW_DATA_FILE, 'r')
        n_samples = f['X'].shape[0]
        print(f"File connected. Total Samples: {n_samples}")
        f.close()
    except FileNotFoundError:
        print("Error: Could not find the raw data file.")
        return

    # Create the Index List
    indices = np.arange(n_samples)
    
    # Shuffle (Reproducible use the same seed each time)
    np.random.seed(2025) 
    np.random.shuffle(indices)

    # Train: 70%
    # Validation: 15%
    # Test: 15%
    n_train = int(0.70 * n_samples)
    n_val = int(0.15 * n_samples)
    
    train_idx = indices[:n_train]
    val_idx = indices[n_train:n_train + n_val]
    test_idx = indices[n_train + n_val:]

    print("\n--- SPLIT REPORT ---")
    print(f"Training Set:   {len(train_idx)} samples (70%)")
    print(f"Validation Set: {len(val_idx)} samples (15%)")
    print(f"Test Set:       {len(test_idx)} samples (15%)")
    print("--------------------")
    
    # Save the Map
    np.savez(SPLIT_MAP_FILE, train=train_idx, val=val_idx, test=test_idx)
    print(f"Success! Master Split Map saved to: {SPLIT_MAP_FILE}")
```

```python
create_master_split()
```
